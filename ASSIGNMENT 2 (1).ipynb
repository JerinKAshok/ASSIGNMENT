# 3. Plot a bar graph of male vs female buyers participated in the sales.

violin plot

```python

sns.violinplot(x='Buyer Gender',y='Buyer Age',hue='New Car',data=c).set(title='Plot of Buyer Gender vs Buyer Age')
```

insights  

*Mean of both male and female members are same in case of  new and old cars  

*Female buyers opt compatrively old cars than new cars

 *Males in between age 30 to 40 opt old cars than new cars

```python
sns.violinplot(x='Buyer Gender',y='Sale Price',hue='New Car',data=c).set(title='Plot of Buyer Gender vs Buyer Age')

```

Insights  

*Mean value of both male and female members are same in case of  new and old cars

*Male Buyers density in between 40000 and 60000 price range is more than female buyers
